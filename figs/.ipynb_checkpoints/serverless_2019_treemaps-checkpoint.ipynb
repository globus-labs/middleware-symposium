```python
# SOURCE: https://mubaris.com/posts/dataviz-treemaps/

# Data Manipulation
import pandas as pd
# Treemap Ploting
import squarify
# Matplotlib and Seaborn imports
import matplotlib
from matplotlib import style
import matplotlib.pyplot as plt
import seaborn as sns
import json
# Activate Seaborn
sns.set()
%matplotlib inline
# Large Plot
matplotlib.rcParams['figure.figsize'] = (16.0, 9.0)
# Use ggplot style
style.use('ggplot')

from pandas.io.json import json_normalize    
```

```python
""" Take the gross JSON-blob and create a more pandas-able CSV"""

with open('/Users/tylerskluzacek/PycharmProjects/GlobusCrawlTest/old_result.json') as data_file:    
    raw_data = json.load(data_file)

csv_file = "mdf_crawled.csv"

import csv
with open(csv_file, "w") as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        writer.writerow(['filepath', 'extension', 'size'])
        for key in raw_data:
            filepath = key
            extension = raw_data[key]["physical"]["extension"]
            size = raw_data[key]["physical"]["size"]
            
            writer.writerow([filepath, extension, size])
```

```python
""" Collect two histograms where key is extension and value is count and size"""

extension_histogram = {}
size_histogram = {}

total_files = 0
total_size = 0
import os 

# print(os.listdir('.'))

# Create histogram of files based on type
with open("mdf_crawled.csv", 'r') as csv_f:
    csv_reader = csv.reader(csv_f, delimiter=',')
    next(csv_reader)  # Skip the header row. 
    for row in csv_reader:
        extension = row[1]
        size = row[2]

#         if extension == 'tiff' or extension == 'tif':
#             print("TIFF!")
#             extension = 'tif/tiff'
        
        if extension not in extension_histogram:
            extension_histogram[extension] = 0
            size_histogram[extension] = 0
            
        extension_histogram[extension] += 1
        size_histogram[extension] += int(size)
        
        total_files += 1
        total_size += int(size)


# print(extension_histogram)
#print(size_histogram)
```

```python
""" Make CSVs for percentage of total files and total size"""

ext_treemap_csv = "ext_mdf_treemap.csv"
size_treemap_csv = "size_mdf_treemap.csv"

with open(ext_treemap_csv, 'w') as g:
    writer = csv.writer(g, delimiter=',')
    writer.writerow(['extension', 'percentage', 'total'])
    for key in extension_histogram:
        writer.writerow([key, int(extension_histogram[key])/total_files, int(extension_histogram[key])]) 
        
with open(size_treemap_csv, 'w') as h:
    writer = csv.writer(h, delimiter=',')
    writer.writerow(['extension', 'size_percentage', 'total'])
    for key in size_histogram:
        writer.writerow([key, int(size_histogram[key])/total_size, int(size_histogram[key])])
```

```python
# Reading CSV file
ext_df = pd.read_csv(ext_treemap_csv)
# Sort by Percentage
ext_df = ext_df.sort_values(by="percentage", ascending=False)

ext_df['label'] = ext_df.apply(lambda row: "{}\n({:0.2f}%)".format(row.extension, row.percentage*100) if row.percentage > 0.0139 else '{}'.format(row.extension) if row.percentage > 0.00145 else '', axis=1)
ext_df['label2'] = ext_df.apply(lambda row: "{}\n({:0.0f})".format(row.extension, row.total) if row.percentage > 0.0139 else '{}'.format(row.extension) if row.percentage > 0.00145 else '', axis=1)


#######

# Reading CSV file
size_df = pd.read_csv(size_treemap_csv)
# Sort by Percentage
size_df = size_df.sort_values(by="size_percentage", ascending=False)
size_df["size_percentage"] = size_df["size_percentage"]*100
size_df = size_df[:250]
size_df['label'] = size_df.apply(lambda row: "{}\n({:0.2f}%)".format(row.extension, row.size_percentage) if row.size_percentage > 1 else '{}'.format(row.extension) if row.size_percentage > 0.0385 else '', axis=1)
size_df['label2'] = size_df.apply(lambda row: "{}\n({:0.2f} TB)".format(row.extension, row.total/(1000000000000)) if row.size_percentage > 1 else '{}'.format(row.extension) if row.size_percentage > 0.0385 else '', axis=1)

size_df
```

```python
# Get Axis and Figure
fig, ax = plt.subplots()

fig.set_size_inches(14, 6)

# Our Colormap  -- available here https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
cmap = matplotlib.cm.GnBu
# Min and Max Values
mini = min(ext_df["percentage"])
maxi = max(ext_df["percentage"])
# Finding Colors for each tile
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in ext_df["percentage"]]

# Plotting: -- NOTE: can set fontweight:'bold' in here. 
squarify.plot(sizes=ext_df["percentage"], label=ext_df["label"], alpha=0.75, color=colors, text_kwargs={'fontsize':10})
# Removing Axis
plt.axis('off')
# Invert Y-Axis
plt.gca().invert_yaxis()
# Title
plt.title("Tree Map: Extensions by Extension Count (% of total)", fontsize=18, fontweight='bold')
# Title Positioning
ttl = ax.title
ttl.set_position([.5, 1.05])

plt.savefig('extension_treemap.jpg', dpi=300)
# BG Color
#fig.set_facecolor('#eeffee')
```
